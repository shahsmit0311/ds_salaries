```python
edu["salary_norm"]=edu["salary_in_usd"]/edu["salary_in_usd"].max()
edu.tail()
```

```python
edu.drop("salary_norm", axis=1, inplace=True)
edu.head()
```

```python
new_row = pd.DataFrame({"work_year": [2022], "experience_level": ["SE"], "employment_type": ["FT"], "job_title": ["Meme Engineer"], "salary": [300000], "salary_currency": ["USD"], "salary_in_usd": [300000], "employee_residence": ["US"], "remote_ratio": [100], "company_location": ["US"],"company_size": ["L"]})
edu = pd.concat([edu, new_row], ignore_index=True)
edu.tail()
```

```python
edu.drop(max(edu.index), axis = 0, inplace = True)
edu.tail()
```

```python
eduDrop = edu.dropna(how = 'any', subset = ["salary_in_usd"])
edu.head()
```

```python
eduFilled = edu.fillna(value = {"salary_in_usd": 0})
eduFilled.head()
```

```python
edu.sort_values(by = 'salary_in_usd', ascending = False, inplace = True)
edu.head()
```

```python
edu.sort_index(axis = 0, ascending = True, inplace = True)
edu.head()
```

```python
group = edu[["job_title","salary_in_usd"]].groupby('job_title').mean()
group.head()
```

```python
filtered_data = edu[edu["work_year"] > 2020]
pivedu = pd.pivot_table(filtered_data , values = 'salary_in_usd', index = ['experience_level'], columns = ['work_year'])
pivedu.head()
```

```python
pivedu.loc[['EN','SE'], [2021,2023]]
```

```python
pivedu = pivedu.dropna()
```

```python
pivedu.rank(ascending = False, method = 'first').head()
```

```python
totalSum = pivedu.sum(axis = 1)
totalSum.rank(ascending = False, method = 'dense').sort_values().head()
```

```python
totalSum = pivedu.sum(axis = 1).sort_values(ascending = False)
totalSum.plot(kind = 'bar', style = 'b', alpha = 0.4, title = "Salary At Different Experience Level")
```

```python
my_colors = ['b', 'r', 'g']
ax = pivedu.plot(kind = 'barh',stacked = True ,color = my_colors)
ax.legend(loc = 'center left', bbox_to_anchor = (1, .5))
```
